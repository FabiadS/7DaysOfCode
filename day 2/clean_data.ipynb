## Dia 02 - Limpeza dos dados

```python
import pandas as pd
```

```python
df_emprestimos_completo = pd.read_csv("./data/df_emprestimos_completo.csv", delimiter=",")
df_emprestimos_completo.head()
```

### Transformações nos dados

#### Excluir a coluna "registro_sistema"

```python
df_emprestimos_completo = df_emprestimos_completo.drop("registro_sistema", axis="columns")
```

```python
df_emprestimos_completo.head()
```

#### Transformar para string a coluna "matricula_ou_siape"

```python
df_emprestimos_completo['matricula_ou_siape'] = df_emprestimos_completo['matricula_ou_siape'].apply(str)
```

```python
df_emprestimos_completo.head()
```

### Adicionar uma nova coluna conforme a localização

```python
class_CDU = []
for localizacao in df_emprestimos_completo['localizacao']:
  if localizacao <= 99:
    class_CDU.append("Generalidades. Ciência e conhecimento")
  elif localizacao >= 100 and localizacao <= 199:
    class_CDU.append("Filosofia e psicologia")
  elif localizacao >= 200 and localizacao <= 299:
    class_CDU.append("Religião")
  elif localizacao >= 300 and localizacao <= 399:
    class_CDU.append("Ciências sociais")
  elif localizacao >= 400 and localizacao <= 499:
    class_CDU.append("Classe vaga. Provisoriamente não ocupada")
  elif localizacao >= 500 and localizacao <= 599:
    class_CDU.append("Matemática e ciências naturais")
  elif localizacao >= 600 and localizacao <= 699:
    class_CDU.append("Ciências aplicadas")
  elif localizacao >= 700 and localizacao <= 799:
    class_CDU.append("Belas artes") 
  elif localizacao >= 800 and localizacao <= 899:
    class_CDU.append("Linguagem. Língua. Linguística")
  else:
    class_CDU.append("Geografia. Biografia. História")

df_emprestimos_completo["CDU"] = class_CDU
```
